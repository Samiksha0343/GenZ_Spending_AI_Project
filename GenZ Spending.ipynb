```python
import pandas as pd

data = pd.read_csv("genz spending.csv")
data
```

```python
data.info()
```

```python
data.describe()
```

```python
import matplotlib.pyplot as plt

plt.bar(data["Age"], data["Income (USD)"])
plt.xlabel("Age")
plt.ylabel("Income (USD)")
plt.title("Age vs Monthly Income")
plt.show()
```

```python
from sklearn.cluster import KMeans

X = data[["Eating Out (USD)", "Online Shopping (USD)", "Savings (USD)"]]

kmeans = KMeans(n_clusters=4)
data["Money_Personality"] = kmeans.fit_predict(X)

data
```

```python
personality_map = {
    0: "Saver",
    1: "Spender",
    2: "Investor",
    3: "Impulsive"
}

data["Personality_Name"] = data["Money_Personality"].map(personality_map)
data
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

X = data.drop(["Money_Personality", "Personality_Name"], axis=1)
y = data["Money_Personality"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model = RandomForestClassifier()
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
import seaborn as sns

sns.scatterplot(
    x=data["Online Shopping (USD)"],
    y=data["Savings (USD)"],
    hue=data["Personality_Name"]
)
plt.xlabel("Online Shopping (USD)")
plt.ylabel("Savings (USD)")
plt.title("Spending vs Savings by Money Personality")
plt.show()
```

```python
data["Online_Shopping_Intensity"] = (
    data["Online Shopping (USD)"] / data["Income (USD)"]
)

sns.boxplot(
    x=data["Personality_Name"],
    y=data["Online_Shopping_Intensity"]
)
plt.title("Online Shopping Intensity by Personality")
plt.show()
```

```python
def recommendation(personality):
    if personality == "Saver":
        return "Good job! You can explore low-risk investments."
    elif personality == "Spender":
        return "Try budgeting apps and reduce unnecessary shopping."
    elif personality == "Investor":
        return "Consider diversifying your investments."
    else:
        return "Track expenses and set monthly spending limits."

data["AI_Recommendation"] = data["Personality_Name"].apply(recommendation)
data
```
